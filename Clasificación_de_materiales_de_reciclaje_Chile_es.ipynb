# Problema

Constantemente generamos desechos que se pueden reciclar, pero clasificarlos en los contenedores de reciclaje o al basurero siguiendo las normas de colores de Chile [Nch 3322 p.27](https://mma.gob.cl/wp-content/uploads/2018/08/Guia-de-Educacion-Ambiental-y-Residuos.pdf) puede ser una tarea dificil.
Esta norma **Nch3322** del Ministerio del Ambiente establece códigos de colores.

Según [Norte Verde](https://norteverdespa.com/blog/reciclaje/el-arcoiris-del-reciclaje-colores-que-hacen-la-diferencia) estos colores consisten en:

- **Color Azul (Papel y cartón)**: e.j. papeles, cartones, cajas o envases de alimentos, periódicos, revistas, papeles de envolver, folletos publicitarios.
- **Color Amarillo (Plásticos)**: e.j. envases y productos plásticos, botellas, envases de alimentos, bolsas. Enjuagar y entregar envases secos.
- **Color Gris Claro (Metales)**: e.j. latas de conserva y refrescos, otros metales. Enjuagados y secados.
- **Color Verde (Vidrio)**: e.j. envases de vidrio, botellas, bebidas y agua. No confudir con cerámico o cristal.
- **Color Beige (Cartón para bebidas)**: e.j. envases de cartón (Tetra Pak) de bebidas, leches, bebidas alcohólicas y alimentos.
- **Color Rojo (Desechos Peligrosos)**: e.j. baterías, pilas, aceites o medicamentos.
- **Color Café (Materia Orgánica)**: e.j. Restos de frutas y verduras, cáscaras de huevo, bolsas de té, pan, arroz, pastas, césped, flores y plantas marchitas.
- **Color Burdeo (Aparatos Eléctricos y Electrónicos)**: e.j. electrodomésticos, dispositivos de audio y video, computadoras.
- **Color Gris Oscuro (Resto de residuos)**: e.j. Que no pueden reciclarse aún.

# Pregunta
¿Con datasets de imágenes de Kaggle, es posible construir un modelo de clasificación de materiales por color de contenedores (según la norma NCH3322)?
# Objetivo
Construir un modelo que clasifique correctamente el color de los contenedores según la norma NCH3322, con un accuracy mínimo del 95%.

# Posibles sesgos (Bias)
- **Sesgo de cobertura (Coverage Bias)**: Puede existir un desbalance en la cantidad de imágenes por tipo de material de reciclaje o desecho. Además, algunos registros podrían incluir ilustraciones o imágenes no representativas del material real, imágenes repetidas, etc. No se consideran detalles específicos como si el envase está enjuagado o seco, ni distinciones precisas como por ejemplo entre vidrio y cristal, lo que podría afectar la representatividad del dataset. 
- **Sesgo de medición (Measurement Bias)**: Pueden presentarse errores o inconsistencias en la asignación de etiquetas a los materiales, ya sea por diferencias en los criterios de clasificación o por errores humanos durante la anotación.

# Análisis de datos

Al momento de realizar una verificación del formato de las imágenes, se ve que pueden utilizar extensiones de tipo `.jpg, .jpeg, .png`

```python
def verify_extensions(path):
    extensiones = Counter()
    for raiz, carpetas, archivos in os.walk(path):
        for archivo in archivos:
            _, extension = os.path.splitext(archivo)
            if extension:
                extensiones[extension.lower()] += 1
    
    for ext, cantidad in extensiones.most_common():
        print(f"{ext}: {cantidad}")
    
    print(f"\nTotal de archivos: {sum(extensiones.values())}")
verify_extensions(MATERIALES_NCH3322_DATASET_PATH)
```

Debido a las diferencias entre los formatos de imagen `.jpg`, `.jpeg` y `.png`, y con el fin de evitar posibles archivos corruptos o incompatibilidades, se decidió convertir todas las imágenes al formato `.jpg` para prevenir inconvenientes futuros.

```python
extensions = ["jpg", "jpeg", "png"]

for label in os.listdir(MATERIALES_NCH3322_DATASET_PATH):
    label_path = os.path.join(MATERIALES_NCH3322_DATASET_PATH, label)
    for filename in os.listdir(label_path):
        file_path = os.path.join(label_path, filename)
        if not any(filename.lower().endswith(ext) for ext in extensions):
            print(f"Eliminando archivo desconocido: {file_path}")
            os.remove(file_path)
            continue
        with PilImage.open(file_path) as img:
            rgb_img = img.convert("RGB")
            new_path = os.path.splitext(file_path)[0] + ".jpg"
            rgb_img.save(new_path, "JPEG")
            if file_path != new_path:
                os.remove(file_path)
print("Conversión completa")
```

Luego se verifican nuevamente las extensiones

```python
verify_extensions(MATERIALES_NCH3322_DATASET_PATH)
```

## Verificar tamaño de las imágenes con `Pillow`
[`Pillow`](https://docs.vultr.com/python/examples/find-the-size-resolution-of-a-image) es una librería basada en `PIL` para abrir, manipular y guardar imágenes con diferentes formatos. Se pueden extraer las dimensiones de ancho y alto. Con su método `Image.open()` se cargan los datos de la imagen y con el método `size` ya se pueden extraer tanto el `width` como el `height`.

A continuación utilizando una muestra de 1000 imágenes se puede ver un gráfico de la distribución de las dimensiones.

# Preprocesamiento de imágenes

## Preprocesamiento en TensorFlow
En [TensorFlow](https://www.tensorflow.org/tutorials/load_data/images#setup) hay al menos 2 formas de pre-procesar imágenes:
- Utilizando utilidades de preprocesamiento de alto nivel de `Keras` con `tf.keras.utils.image_dataset_from_directory` y capas `tf.keras.layers.Rescaling`.
- Creando un flujo de entrada propio desde cero con `tf.data`.
Según la documentación de TensorFlow, en general lo más conveniente es el preprocesamiento con las utilidades de `Keras`, pero para un control más fino o granular, se puede crear un `pipeline` con `tf.data`.

## Diccionarios para etiquetas
Dado que las variables de las etiquetas son de tipo `string`, luego será necesario convertirlas a un tipo numérico.
Por lo que es útil crear diccionarios para mapear entre la etiqueta y su correspondiente número.

Entonces, al contenedor `Color_amarillo` le correspondería el `0`, al contenedor `Color_azul` el `1` y así sucesivamente.

```python
image_range = range(len(labels))
int_to_label  = dict(zip(image_range, labels))
label_to_int  = dict(zip(labels, image_range))
print(label_to_int)
print(int_to_label)
```

## Dividir las imágenes y etiquetas del dataset en subconjuntos
A continuación se define una función para extraer imágenes y etiquetas ya divididas en subconjuntos de datos para `train`, `valid` y `test`.
Se aclara que debido al desbalance de los datos, la división de los datos es a nivel de etiquetas para conservar una muestra más balanceada entre subconjuntos.

Entonces en `train_test_valid_split` por cada imagen de una etiqueta se van creando listas con la ruta de la imagen y la etiqueta a la que pertenece, luego se estima la cantidad de elementos para cada subconjunto y finalmente mediante `slicing` se agrega cierta cantidad de `image_items` en cada subconjunto.

```python
VALID_SIZE=0.1
TEST_SIZE=0.1
def train_test_valid_split(test_size=TEST_SIZE, valid_size=VALID_SIZE):
    train = []
    valid = []
    test = []
    for label in labels:
        images = os.listdir(f"{MATERIALES_NCH3322_DATASET_PATH}/{label}")
        image_items = [(f"{MATERIALES_NCH3322_DATASET_PATH}/{label}/{i}", label_to_int[label]) for i in list(images)]
        n_total = len(images)
        n_test = int(TEST_SIZE * n_total)
        n_valid = int(VALID_SIZE * n_total)

        train.extend(image_items[n_test+n_valid:])
        valid.extend(image_items[n_test:n_test+n_valid])
        test.extend(image_items[:n_test])

    X_train, y_train = zip(*train)
    X_test, y_test = zip(*test)
    X_valid, y_valid = zip(*valid)
    return list(X_train), list(X_test), list(X_valid), list(y_train), list(y_test), list(y_valid)
```

Luego se llama a la función `train_test_valid_split`, se imprime la cantidad de elementos por cada subconjunto y el elemento de índice 3 del subconjunto `train`.

```python
X_train, X_test, X_valid, y_train, y_test, y_valid = train_test_valid_split()
print(len(X_train), len(X_valid), len(X_test))
print(X_train[3], y_train[3])
```

## Creación de Datasets de Tensorflow para el procesamiento de imágenes
Mediante `tf.data.Dataset.from_tensor_slices` se crean datasets para los subconjuntos `train`, `valid` y `test`. Luego se le aplica un `shuffle` al dataset de entrenamiento `train_ds` para evitar sesgos en el entrenamiento.

```python
train_ds = tf.data.Dataset.from_tensor_slices((X_train, y_train))
valid_ds = tf.data.Dataset.from_tensor_slices((X_valid, y_valid))
test_ds = tf.data.Dataset.from_tensor_slices((X_test, y_test))
train_ds = train_ds.shuffle(train_ds.cardinality())
```

## Preprocesamiento de imágenes
El procesamiento empieza con la definición de una función [`parse_image`](https://www.tensorflow.org/guide/data#:~:text=def%20parse_image(filename)%3A) encargada de recibir la ruta de la imagen `filename` y el correspondiente `label`, luego se leerá, decodificará con canales RGB, se normalizará y convertirá a datos numéricos, por último se ajustará su tamaño y se retornará una tupla con la imagen ya procesada y la etiqueta a la que pertenece.
> Se decide definir `IMG_HEIGHT=128` y `IMG_WIDTH=128` se basa en adoptar un tamaño estándar y comunmenete utilizado en redes `CNN`. Luego también se puede cambiar por `IMG_HEIGHT=224` y `IMG_WIDTH=224` o usando `parse_image_224` para tratar de obtener un mejor resultado.

```python
IMG_HEIGHT = 128
IMG_WIDTH = 128

def parse_image(filename, label):
    image = tf.io.read_file(filename)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.convert_image_dtype(image, tf.float32)
    image = tf.image.resize(image, [IMG_HEIGHT, IMG_WIDTH])    
    return image, label
def parse_image_224(filename, label):
    image = tf.io.read_file(filename)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.convert_image_dtype(image, tf.float32)
    image = tf.image.resize(image, [224, 224])    
    return image, label
```

A continuación se realiza una breve demostración para ver el uso de `parse_image` con unas muestras del dataset `train_ds`.
Lo que se espera es que muestre imágenes ya preprocesadas de `parse_image`, luego con el método `imshow` muestre las imágenes y adjunte como titulo la etiqueta a la cual pertence.

```python
def show(image, label):
    plt.figure()
    plt.imshow(image)
    plt.title(int_to_label[label])
    plt.axis('off')
    plt.show();
for filename, label in train_ds.take(2):
    image, label = parse_image(filename, label)
    show(image.numpy(), label.numpy())
```

A continuación, definimos la función `augment_image`, diseñada para aumentar los datos de imágenes durante el proceso de entrenamiento. Como señala [Kalana Hewapathirana](https://medium.com/@kalanarandil/%EF%B8%8F-data-augmentation-in-machine-learning-flip-rotate-zoom-fe33dbd7c324) en un post de **Data Augmentation**
, existen diversas formas de realizar data augmentation. Una de ellas es la augmentación personalizada utilizando `tf.image`, sin necesidad de generar nuevas imágenes, ofrece un control total sobre las transformaciones y adecuada para pipelines de entrenamiento.

```python
def augment_image(image, label):
    image = tf.image.random_flip_left_right(image)
    image = tf.image.random_flip_up_down(image)
    image = tf.image.random_brightness(image, max_delta=0.1)
    image = tf.image.random_contrast(image, 0.8, 1.2)
    return image, label
```

Luego, según la [documentación de TensorFlow](https://www.tensorflow.org/guide/data#preprocessing_data) ya se podría preprocesar las imágenes con `parse_image` y `augment_image` mediante el método `map`.
Después, se utiliza `batch` para poder dividir los elementos de cada dataset en partes que se van procesar y entrenar en paralelo.
Depués según [sparkcodehub](https://www.sparkcodehub.com/tensorflow/data-handling/how-to-prefetch-datasets#google_vignette) existe una optimización importante que se puede hacer, que es el `prefetching`, Esto prepara el siguiente batch en la CPU mientras la GPU entrena el batch actual, reduciendo los cuellos de botella y el tiempo muerto durante el entrenamiento.

En este caso, se selecciona un tamaño de batch de 32, es decir, 32 elementos a procesar en paralelo en cada iteración. Como es probable que el último batch no cumpla con ese tamaño, se puede usar `drop_remainder=True` para ignorar el último batch incompleto.

```python
BATCH_SIZE=32
AUTOTUNE = tf.data.AUTOTUNE

train_ds = (
    train_ds
    .map(parse_image, num_parallel_calls=AUTOTUNE)
    .map(augment_image, num_parallel_calls=AUTOTUNE)
    .batch(BATCH_SIZE, drop_remainder=True)
    .prefetch(AUTOTUNE)
)

valid_ds = (
    valid_ds
    .map(parse_image, num_parallel_calls=AUTOTUNE)
    .batch(BATCH_SIZE)
    .prefetch(AUTOTUNE)
)

test_ds_224 = (
    test_ds
    .map(parse_image_224, num_parallel_calls=AUTOTUNE)
    .batch(BATCH_SIZE)
    .prefetch(AUTOTUNE)
)


test_ds = (
    test_ds
    .map(parse_image, num_parallel_calls=AUTOTUNE)
    .batch(BATCH_SIZE)
    .prefetch(AUTOTUNE)
)
```

# Construcción de Modelos

## Construcción de Modelo CNN (Convolutional Neural Network)
El primer modelo a utilizar será un modelo básico de [`CNN`](https://arxiv.org/pdf/1511.08458), que servirá como modelo `baseline`.
Los modelos `CNN` son ampliamente utilizados para la clasificación de imágenes, ya que permiten extraer características locales de las imágenes, a diferencia de un modelo `MLP` (Multi-Layer Perceptron), que solo identifica patrones de forma global.
Existen varias razones por las que un MLP no es ideal para estos casos, entre ellos es que las `CNN` son invariantes a traslaciones (translation-invariant) y que un `MLP` aplicado directamente a imágenes puede volverse muy complejo y propenso a overfitting.

Para construir un modelo `CNN`, TensorFlow proporciona una serie de métodos para crear capas tales como `Conv2D`, `MaxPooling2D`, `Flatten` y `Dense`. Una arquitectura de referencia puede verse en [datacamp](https://www.datacamp.com/tutorial/cnn-tensorflow-python) , donde el cuerpo del modelo (`body`) utiliza 2 bloques de capas de convolución y pooling. Posteriormente, los mapas de activación se transforman mediante `Flatten` para adaptarlos a la parte final del modelo (`head`) en la cual las capas `Dense` se encargan de la detección de patrones y la última capa produce las probabilidades asociadas a cada etiqueta.

```python
from tensorflow.keras import layers, models
INPUT_SHAPE = (128, 128, 3)
FILTER1_SIZE = 32
FILTER2_SIZE = 64
FILTER_SHAPE = (3, 3)
POOL_SHAPE = (2, 2)
FULLY_CONNECTED_NUM = 128
NUM_CLASSES = len(labels)

model = models.Sequential([
    layers.Input(shape=INPUT_SHAPE),
    layers.Conv2D(FILTER1_SIZE, FILTER_SHAPE, activation='relu'),
    layers.MaxPooling2D(POOL_SHAPE),
    layers.Conv2D(FILTER2_SIZE, FILTER_SHAPE, activation='relu'),
    layers.MaxPooling2D(POOL_SHAPE),
    layers.Flatten(),
    layers.Dense(FULLY_CONNECTED_NUM, activation='relu'),
    layers.Dense(NUM_CLASSES, activation='softmax')
])
```

A continuación, se realiza la compilación del modelo. Para ello, se utiliza el optimizador `Adam`, encargado de aplicar el algoritmo de decenso de gradiente durante el entrenamiento.
Se define la función de pérdida `sparse_categorical_crossentropy`, adecuada cuando las etiquetas se encuentran codificadas mediante `Integer Encoding`.
Finalmente, se incorpora la métrica `accuracy` para evaluar el desempeño del modelo durante el entrenamiento y validación.

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

Luego se procede a entrenar el modelo

```python
import random
import numpy as np
import tensorflow as tf
SEED_VALUE = 42
random.seed(SEED_VALUE)
np.random.seed(SEED_VALUE)
tf.random.set_seed(SEED_VALUE)
history = model.fit(
    train_ds,
    validation_data=valid_ds,
    epochs=20
)
model.save("basic_cnn.keras")
import pickle
pickle.dump(history.history, open("history_basic_cnn.pkl", "wb"))
```

```python
import pickle
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

history = pickle.load(open("history_basic_cnn.pkl","rb"))
train_loss = history["loss"]
valid_loss = history["val_loss"]
train_acc = history["accuracy"]
valid_acc = history["val_accuracy"]

def plot_metrics(metrics, metric_name=None, title=None, ylim=None, ylabel=None, color=None):
    fig, ax = plt.subplots(figsize=(15,5))

    if not isinstance(metric_name, (list, tuple)):
        metric_name = [metric_name,]
        metrics = [metrics,]

    for idx, metric in enumerate(metrics):
        ax.plot(metric, color=color[idx])

    plt.title(title)
    plt.xlim([0, len(metrics[0])-1])
    if ylim:
        plt.ylim(ylim)
    plt.xlabel("Epochs")
    plt.ylabel(ylabel)
    plt.legend(metric_name)
    ax.xaxis.set_major_locator(MultipleLocator(5))
    ax.xaxis.set_major_formatter(FormatStrFormatter("%d"))
    ax.xaxis.set_minor_locator(MultipleLocator(1))
    plt.grid(True)
    plt.savefig(f"metric-{title}.jpg", dpi=300, bbox_inches='tight')
    plt.show()

plot_metrics(
    [train_loss, valid_loss],
    metric_name=["train_loss", "val_loss"],
    title="Epochs vs Loss",
    ylabel="Loss",
    ylim=[0, max(max(train_loss), max(valid_loss)) + 0.1],
    color=["g", "b"]
)

plot_metrics(
    [train_acc, valid_acc],
    metric_name=["train_acc", "val_acc"],
    title="Epochs vs Accuracy",
    ylabel="Accuracy",
    ylim=[0.4, 1],
    color=["g", "b"]
)
```

## Aplicar transfer learning con [MobileNetV2](https://keras.io/api/applications/mobilenet)
Modelos como MobileNetv2 son especialmente adecuados para aplicaciones de visión por computadora en dispositivos móviles, ya que combinan buena precisión con bajo costo computacional.
Utilizando el método `MobileNetV2`creamos una base del modelo sin incluir la parte `head` de clasificación final, porque justamente esa parte se va a adaptar a nuestras clases.
Luego congelamos los pesos de este modelo base con `base_model.trainable=False`para aprovechar estos en la creación de un nuevo modelo enfocado al problema principal.
Entonces crearemos un nuevo modelo `model` incluyendo `base_model` y la nueva parte `head` en que consideraremos las clases correspondientes a los colores de los contenedores.
Finalmente se agregan 2 callbacks `early_stop` para que el entrenamiento se detenga una vez el modelo deje de mejorar en `val_loss` y `checkpoint`para guardar el mejor modelo obtenido durante el entrenamiento, luego se entrena el modelo con 10 `epochs`.

```python
import random
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

SEED_VALUE=42
random.seed(SEED_VALUE)
np.random.seed(SEED_VALUE)
tf.random.set_seed(SEED_VALUE)

IMG_HEIGHT = 128
IMG_WIDTH = 128
BATCH_SIZE = 32
NUM_CLASSES = len(labels)
AUTOTUNE = tf.data.AUTOTUNE

base_model = MobileNetV2(
    input_shape=(IMG_HEIGHT, IMG_WIDTH, 3),
    include_top=False,
    weights="imagenet"
)
base_model.trainable = False

model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(NUM_CLASSES, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)
checkpoint = ModelCheckpoint(
    "mobilenetv2_best.keras",
    monitor='val_accuracy',
    save_best_only=True
)

history = model.fit(
    train_ds,
    validation_data=valid_ds,
    epochs=10,
    callbacks=[early_stop, checkpoint]
)

import pickle
pickle.dump(history.history, open("history_mobilenetv2_best.pkl", "wb"))
```

## Comparación entre modelos

A modo de comparar la generalización de los modelos, se crea una función `get_scores` en que por cada modelo, va capturando el `score` obtenido.

```python
from tensorflow.keras.models import load_model
import pandas as pd

cnn = load_model("basic_cnn.keras")
mobilenetv2_best = load_model("mobilenetv2_best.keras")
mobilenetv2_best_224 = load_model("mobilenetv2_best_224.keras")

models = {
    "Básico CNN": cnn, 
    "MobilenetV2": mobilenetv2_best, 
    "MobilenetV2 224": mobilenetv2_best_224
}
def get_scores(models):
    scores = {}
    for name, model in models.items():
        _ , acc = model.evaluate(test_ds if name != "MobilenetV2 224" else test_ds_224)
        scores[name] = acc
    return scores
scores = get_scores(models)
```

Luego se muestra en un gráfico los resultados del accuracy de cada modelo.

```python
print(scores)
df = pd.DataFrame(scores, index=["accuracy"])
df.T.plot.bar()
plt.xticks(rotation=0)
plt.grid(True)
plt.ylim([0.83,1])
plt.savefig("accuracy_scores.png")
plt.show();
```

Considerando el gráfico anterior, se toma la decisión de continuar con el tercer modelo `MobilenetV2 224` relacionado a la mejora en el aumento de tamaño de imagen a 224.

## Predicciones
Previo a la evaluación del modelo mediante métricas, se realizan predicciones con `test_ds_224` puesto que el modelo necesita imágenes de tamaño 224.
Para un mejor manejo de imágenes y etiquetas, se extraen `X_test` y `y_test` de `test_ds_224`. Y luego se obtienen predicciones en `y_preds`.

```python
X_test = []
y_test = []
for x, y in test_ds_224:
    X_test.extend(x.numpy())
    y_test.extend(y.numpy())
y_preds = mobilenetv2_best_224.predict(test_ds_224)
```

Para visualizar las predicciones, se selecciona una muestra aleatoria de 15 imágenes a mostrar, luego con `imshow` se muestran las imágenes, un titulo con la predicción (P) y el valor real (R), finalmente en caso de que sean iguales, el titulo se muestra con el color `green` y si son diferentes, del color `red`.

```python
import matplotlib.pyplot as plt
import numpy as np
import random

n = 15
cols = 5
rows = int(np.ceil(n / cols))

plt.figure(figsize=(cols * 3, rows * 3))

indices = random.sample(range(len(y_preds)), n)

for plot_i, idx in enumerate(indices):
    plt.subplot(rows, cols, plot_i + 1)
    plt.imshow(X_test[idx])
    
    p = y_preds[idx]
    pred_label = int_to_label[np.argmax(p)]
    true_label = int_to_label[y_test[idx]]
    color = "green" if pred_label == true_label else "red"
    
    plt.title(f"P: {pred_label}\nR: {true_label}", color=color, fontsize=9)
    plt.axis("off")

plt.tight_layout()
plt.show()
```

Esto nos muestra que el modelo no es perfecto, se visualizar que tiene dificultades en diferenciar entre desechos orgánicos y vidrio (`Color_cafe` y `Color_verde`). También en diferenciar entre metales y vidrio (`Color_gris_claro` y `Color_verde`).

## Matriz de Confusión

Según [datacamp](https://www.datacamp.com/es/tutorial/what-is-a-confusion-matrix-in-machine-learning), permite una visión más profunda sobre el rendimiento, errores y puntos débiles del modelo.

Se utiliza `seaborn` para mostrar un mapa de calor de la matriz de confusión creada de `tf.math.confusion_matrix`, dado a que las predicciones de `y_preds` entregarán probabilidades, se obtendrá el índice de la máxima probabilidad de cada predicción, y como los datos de `test_ds` estan en `Integer_Encoding` se listan en `y_test`.

```python
import seaborn as sn

y_test = []
for _, y in test_ds:
    y_test.extend(y.numpy())

y_preds_max = [np.argmax(i) for i in y_preds]

cm = tf.math.confusion_matrix(y_test, y_preds_max)
df_cm = pd.DataFrame(cm.numpy(), 
                     index=int_to_label.values(), 
                     columns=int_to_label.values())

fig, ax = plt.subplots(figsize=(8,4))
ax = sn.heatmap(df_cm, annot=True, fmt="d", annot_kws={"size":14}, cmap="YlGnBu")
plt.xlabel("Valores reales")
plt.ylabel("Valores predicción")
plt.savefig("confusion_matrix.png")
plt.show();
```

- `Color_amarillo`: Se clasificaron correctamente 166 imágenes de materiales de reciclaje. Hubo 33 predicciones incorrectas, principalmente confundidas con `Color_gris_claro` y `Color_verde`.
- `Color_burdeo`: Se clasificaron correctamente 244 imágenes. Hubo 26 predicciones erróneas, la mayoría confundidas con `Color_rojo`.
- `Color_gris_claro`: Se clasificaron correctamente 161 imágenes. Hubo 35 errores de clasificación, principalmente hacia `Color_amarillo` y `Color_verde`.
- `Color_rojo`: Se clasificaron correctamente 320 imágenes. Hubo 32 clasificaciones incorrectas, principalmente hacia `Color_burdeo`.
- `Color_verde`: Se clasificaron correctamente 165 imágenes. Se realizaron 36 predicciones erróneas, mayormente confundidas con `Color_gris_claro`. Además, 23 imágenes reales de `Color_verde` fueron clasificadas erróneamente en otros contenedores, principalmente `Color_amarillo` y `Color_gris_claro`.

A partir del análisis de la matriz de confusión, se observa que, de los nueve colores de contenedores, cinco `Color_amarillo`, `Color_gris_claro`, `Color_verde`, `Color_burdeo` y `Color_rojo` presentaron cierta dificultad en la clasificación. 

En resumen se muestra una confusión moderada en que el modelo no distingue perfectamente pero tampoco falla de forma severa, para mejorar el rendimiento del modelo, sería recomendable analizar las causas de dichas confusiones, similitudes visuales entre clases, aumentar el número de muestras en las clases más problemáticas y verificar posibles errores de etiquetado manual entre colores.
Luego se continúa la evaluación con el reporte de clasificación `classification_report`.

## Métricas de `classification_report`

Resume el rendimiento del modelo, considerando métricas como `precision`, `recall`, `f1-score`.

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_preds_max, target_names=int_to_label.values()))
```

También es posible detectar dificultades en las 5 clases de colores antes mencionadas (`Color_amarillo`, `Color_gris_claro`, `Color_verde`, `Color_burdeo` y `Color_rojo`) ya sea por parte de `precision` o `recall`.
Luego con `f1-score` hay clases como `Color_verde`, `Color_gris_claro`, `Color_amarillo` cerca del 0.8, esto indica que hay un considerable desbalance entre `precision` y `recall` que dificulta una buena clasificación. Esto es útil porque mejorando de alguna manera el modelo para las clases `Color_verde`, `Color_gris_claro`, `Color_amarillo` se estaría también mejorando significativamente el `f1-score` y por ende una mejor clasificación.

Respecto al accuracy, si bien el accuracy de `test_ds` es de 0.9370192289352417, `classification_report` lo redondea a 0.94.
Tanto para el promedio ponderado y no ponderado de las métricas ninguna de ellas esta bajo del 0.9, por lo que ciertamente si bien hay una cierta dificultad de clasificación, es posible seguir mejorando el modelo.
